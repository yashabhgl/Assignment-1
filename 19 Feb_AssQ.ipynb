```python
#Q3. What is App routing in Flask? Why do we use app routes?

#Ans.
"""
In Flask, app routing refers to the process of defining URL patterns (or routes) for your web application. 
These URL patterns define how the user can interact with your application and what content or functionality 
will be displayed based on the user's requests.

App routing in Flask is implemented using the @app.route decorator, which is used to bind a function to a URL route.
The decorator takes a URL pattern as an argument, and when a user requests that URL,
Flask will call the corresponding function and return the response to the user.

Example- 

from flask import Flask

app = Flask(__name__)

@app.route('/')
def home():
    return 'Hello, world!'
    
    
App routing in Flask is useful for creating a clean and structured application that is easy to navigate and understand. 
It allows you to define different routes for different functionality in your application and 
helps you organize your code in a more modular and maintainable way.

In summary, we use app routes in Flask to define URL patterns and map them to functions that handle user requests. 
This helps us create a structured and modular application that is easy to navigate and maintain.    
"""
```

```python
"""
Q4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/”
route to show the following details:
Company Name: ABC Corporation
Location: India
Contact Detail: 999-999-9999
Attach the screenshot of the output in Jupyter Notebook.
"""

#Ans.

from flask import Flask

app = Flask(__name__)

@app.route("/")
def hello_world():
    return "<h1>Company Name: ABC Corporation</h1>\n<h1>Location: India</h1>\n<h1>Contact Detail: 999-999-9999</h1>"

@app.route("/welcome")
def test():
    return "<h2>Welcome to ABC Corporation</h2>"


if __name__=="__main__":
    app.run(host="0.0.0.0")
```

```python
#Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

#Ans.
"""
In Flask, the url_for() function is used for URL building.
This function generates a URL for the given endpoint with the specified arguments.

The url_for() function takes the name of the endpoint as its first argument and any number of keyword arguments as its remaining arguments. 
The keyword arguments represent the variable parts of the URL and their values.
The function then returns a URL for the specified endpoint with the variable parts replaced by their values.

The url_for() function is especially useful when you need to generate URLs dynamically in your templates or code, 
and it ensures that the generated URLs are always correct and up-to-date, 
even if you change the URL structure or endpoint names in your application.
"""

# Code for url_for() function.
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/user/<username>')
def profile(username):
    return f"Hello, {username}!"

with app.test_request_context():
    print(url_for('profile', username='Yashabh'))

"""
Here, the url_for() function is used to generate a URL for the profile endpoint with the value of the username variable set to Yashabh. 
The output of the print() statement will be /user/Yashabh, 
which is the URL for the profile endpoint with the username variable set to Yashabh.
"""
```
